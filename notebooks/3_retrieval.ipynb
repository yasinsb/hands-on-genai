## Creating a collection by adding docs

```python

## Add documents
from dbs.vectordb.qdrant_client import add_documents, clean_collection, search_documents
```

```python
# make sure to clean the collection first
clean_collection("my_collection")
```

```python
docs = [
    "France is a country in Western Europe.",
    "Germany is a country in Central Europe.",
    "Italy is a country in Southern Europe.",
]

add_documents("my_collection", docs)
```

```python
points = search_documents(collection_name="my_collection", query="France", k=10)
```

```python
len(points)
```

```python
for point in points:
    print(f"Document ID: {point.id}")
    print(f"Document Score: {round(point.score, 2)}")
    print(f"Document Text: {point.document}")
    # print(f"Document Metadata: {doc.document.metadata}")s
    print("-" * 100)
    
```
