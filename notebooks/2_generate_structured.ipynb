```python
from pydantic import BaseModel, Field
```

## Define prompts and output structure

```python
class MyModel(BaseModel):
    capital: str = Field(description="The capital city of the given country")

system_prompt = "You are a helpful assistant."
user_prompt = "France"
```

## Generate with Gemini

```python

from genai_clients.structured.gemini import generate_structured as generate_structured_gemini
```

```python
output = generate_structured_gemini(system_prompt, user_prompt, MyModel)
print(output)  # MyModel(capital='Paris')
```

```python
from genai_clients.structured.gpt import generate_structured as generate_structured_gpt
```
