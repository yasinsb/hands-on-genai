## Generate with Gemini

```python
from genai_clients.json.gemini import generate_json as generate_json_gemini
```

```python
system_prompt = "You are a helpful assistant."
user_prompt = "What is the capital of France?"

response_schema = {
    "type": "object",
    "properties": {
        "capital": {"type": "string"}
    },
    "required": ["capital"]
}

output = generate_json_gemini(system_prompt, user_prompt, response_schema)

print(output)
```

```python
# Now the output is a dictionary!
print(output["capital"])
```
